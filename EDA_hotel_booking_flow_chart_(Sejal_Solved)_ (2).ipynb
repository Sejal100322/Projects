### Variables Description

*   hotel - **Type of hotel(Resort or City).**

* stays_in_week_nights -**No. of weeknight( Mon to Fri) spend by guest at hotel.**
* adults - **No. of adults among guest.**
* children - **No. of children among guest.**
* babies - **No. of babies among guest.**
* meal - **Type of meal booked by guest.**
* country - **Country of guest.**
* market_segment -  **grouping into categories based on their booking patterns and travel habits.**
* distribution_channel - **Name of booking distribution channel.**
* is_repeated_guest - **If the booking was from repeated by guest(1) or not(0).**
* previous_cancelletions - **No. of previous booking that were cancelled by the customer prior to the customer booking.**
* previous_booking_not_canceled - **No. of previous booking that were not cancelled by the customer prior to the customer booking.**
* reserved_room_type - **Code of room type reserved.**
* assigned_room_type - **code of room type reserved.**
* booking_changes - **No. of changes made to the booking.**

* company - **Id of the company that made the booking.**
* days_in_waiting_list - **No. of days the booking on the waiting list.**

* adr - **(Average Daily Rate) dividing the sum of lodgging transaction by the total number of staying night.**

### Chart - 7
####*Question no. 7 What is the pecentage of Booking cancellation for both type*

```python
# Group by 'hotel' and 'is_canceled', then count the number of cancellations for each group
cancellation_counts = df1.groupby(['hotel', 'is_canceled']).size().reset_index(name='booking_count')

# Pivot the data to separate cancellations (1) and non-cancellations (0)
cancellation_pivot = cancellation_counts.pivot(index='hotel', columns='is_canceled', values='booking_count').reset_index()

# Rename columns for better understanding
cancellation_pivot.columns = ['hotel', 'non_canceled', 'canceled']

# Calculate the percentage of cancellations for each hotel
cancellation_pivot['cancellation_percentage'] = (cancellation_pivot['canceled'] / (cancellation_pivot['canceled'] + cancellation_pivot['non_canceled'])) * 100

# Display the result
print(cancellation_pivot[['hotel', 'cancellation_percentage']])
```

```python
# Plotting the cancellation percentages
plt.bar(cancellation_pivot['hotel'], cancellation_pivot['cancellation_percentage'], color=['red', 'blue'])
plt.xlabel('Hotel')
plt.ylabel('Cancellation Percentage')
plt.title('Booking Cancellation Percentage for Hotels')
plt.show()
```

#Insights

**City Hotel:** Higher cancellation rate (30.1%) compared to Resort Hotel (23.5%).
Reason: City Hotel may have more transient customers or lenient policies, leading to higher cancellations.
Impact: Affects revenue predictability; stricter policies or discounts could help.

**Resort Hotel:** Lower cancellations, likely due to leisure-oriented, pre-planned bookings.

```python
# Calculate the percentage of repeated guests for Resort Hotel
resort_hotel_data = df1[df1['hotel'] == 'Resort Hotel']
resort_repeated_percentage = (resort_hotel_data['is_repeated_guest'].sum() / len(resort_hotel_data)) * 100

# Calculate the percentage of repeated guests for City Hotel
city_hotel_data = df1[df1['hotel'] == 'City Hotel']
city_repeated_percentage = (city_hotel_data['is_repeated_guest'].sum() / len(city_hotel_data)) * 100

# Print the results
print(f"Repeated customer percentage for Resort Hotel: {resort_repeated_percentage:.2f}%")
print(f"Repeated customer percentage for City Hotel: {city_repeated_percentage:.2f}%")

# Data for the bar chart
hotel_types = ['Resort Hotel', 'City Hotel']
repeated_percentages = [resort_repeated_percentage, city_repeated_percentage]

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(hotel_types, repeated_percentages, color=['#4CAF50', '#2196F3'])

# Adding labels and title
plt.title('Repeated Customer Percentage by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Repeated Customer Percentage (%)')

# Show the plot
plt.show()
```

### Chart - 14
####*Question no. 14 What is optimal stay length in both types of hotel ie. City hotels and Resort Hotels*

```python
# Calculate the total length of stay by summing stays_in_week_nights and stays_in_weekend_nights
df1['total_stay_length'] = df1['stays_in_week_nights'] + df1['stays_in_weekend_nights']

# Group by hotel type (Resort Hotel and City Hotel) and calculate the mode (most common stay length)
resort_hotel_data = df1[df1['hotel'] == 'Resort Hotel']
city_hotel_data = df1[df1['hotel'] == 'City Hotel']

# Find the most common (mode) stay length for Resort Hotel
resort_optimal_stay = resort_hotel_data['total_stay_length'].mode()[0]

# Find the most common (mode) stay length for City Hotel
city_optimal_stay = city_hotel_data['total_stay_length'].mode()[0]

# Print the results
print(f"Optimal stay length for Resort Hotel: {resort_optimal_stay} nights")
print(f"Optimal stay length for City Hotel: {city_optimal_stay} nights")
```

- **Resort Hotel** has an optimal stay length of **1 night**, suggesting that guests typically stay for short stays, possibly for weekend getaways or quick breaks.
- **City Hotel** has an optimal stay length of **3 nights**, likely reflecting business or leisure trips where guests stay a bit longer.

In summary, City Hotels tend to have slightly longer stays compared to Resort Hotels.

####*ADR relationship with total number of people*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming df1 is already loaded

# Calculate the total stay length
df1['total_stay_length'] = df1['stays_in_week_nights'] + df1['stays_in_weekend_nights']

# Calculate the total number of people
df1['total_people'] = df1['adults'] + df1['children'].fillna(0) + df1['babies']

# Print summary of the calculated columns
print("Total Stay Length Summary:")
print(df1['total_stay_length'].describe())
print("\nTotal People Summary:")
print(df1['total_people'].describe())

# Create subplots for better comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1. Relationship between ADR and Total Stay Length
sns.regplot(x='total_stay_length', y='adr', data=df1, scatter_kws={'s': 20, 'alpha': 0.5, 'color': 'blue'},
            line_kws={'color': 'red'}, ax=axes[0])
axes[0].set_title('Relationship between ADR and Total Stay Length')
axes[0].set_xlabel('Total Stay Length (nights)')
axes[0].set_ylabel('ADR (Average Daily Rate)')
axes[0].grid(True)

# 2. Relationship between ADR and Total Number of People
sns.regplot(x='total_people', y='adr', data=df1, scatter_kws={'s': 20, 'alpha': 0.5, 'color': 'green'},
            line_kws={'color': 'red'}, ax=axes[1])
axes[1].set_title('Relationship between ADR and Total Number of People')
axes[1].set_xlabel('Total Number of People')
axes[1].set_ylabel('ADR (Average Daily Rate)')
axes[1].grid(True)

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

#### Chart - 17 - Correlation Heatmap

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataset is already loaded in the 'df1' DataFrame

# Select only numeric columns for correlation calculation
numeric_df1 = df1.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numeric columns
correlation_matrix = numeric_df1.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='RdYlBu',  # Choosing a more intuitive color palette
            fmt='.2f',
            linewidths=0.5,
            mask=mask,  # Mask the upper triangle
            vmin=-1, vmax=1,  # Set the color scale limits for correlation
            annot_kws={'size': 10},  # Adjust annotation font size
            cbar_kws={'label': 'Correlation Coefficient'},  # Colorbar label
            square=True,  # Ensures the heatmap is square
            linecolor='white',  # Set line color between cells
            cbar=True)

# Set the title and show the plot
plt.title('Correlation Heatmap')
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```
